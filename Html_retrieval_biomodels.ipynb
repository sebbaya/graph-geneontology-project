```python
import os
import zipfile
import requests
import json
from bioservices import BioModels
s = BioModels()
```

```python
def get_all_models(query, page_size=10):
    """
    Fonction pour récupérer tous les modèles correspondant à la requête avec pagination.
    """
    models = []
    offset = 0
    s = BioModels()

    while True:
        try:
            search_results = s.search(query, offset=offset)

            if 'models' not in search_results or not search_results['models']:
                break

            models.extend(search_results['models'])
            offset += page_size
            print(f"Page {offset // page_size} téléchargée, {len(search_results['models'])} modèles récupérés.")

        except Exception as e:
            print(f"Erreur lors de la récupération des modèles : {e}")
            break

    return models
print("dengue:", (get_all_models('dengue AND curationstatus:"Manually curated" AND modelformat:"SBML" AND TAXONOMY:9606')))
print("west nile:", (get_all_models('"west nile" AND curationstatus:"Manually curated" AND modelformat:"SBML"AND TAXONOMY:9606')))
print("chikungunya:", (get_all_models('chikungunya AND curationstatus:"Manually curated" AND modelformat:"SBML"AND TAXONOMY:9606')))
print("mpox:", (get_all_models('(mpox OR monkeypox) AND curationstatus:"Manually curated" AND modelformat:"SBML"AND TAXONOMY:9606')))
print("avian influenza:", (get_all_models('"avian influenza" AND curationstatus:"Manually curated" AND modelformat:"SBML"AND TAXONOMY:9606')))
print("covid:", (get_all_models('(covid OR "SARS-CoV-2") AND curationstatus:"Manually curated" AND modelformat:"SBML"AND TAXONOMY:9606')))
```

```python
def download_model_file(model_id, sbml_url, directory, curation, disease):
    try:
        sbml_filename = f"{model_id}.xml"
        model_path = os.path.join(directory, sbml_filename)

        response = requests.get(sbml_url)
        print(response.headers.get("Content-Type"), response.text[:30])
        if response.status_code == 200:
            with open(model_path, 'wb') as f:
                f.write(response.content)
            print("OK", curation, disease, model_id)
        else:
            raise RuntimeError(f"Erreur lors du téléchargement du modèle {model_id}: {response.status_code}")

        return model_path

    except Exception as e:
        print(f"Erreur lors du téléchargement du modèle {model_id}: {e}")
        return None
def download_model_with_metadata(model_data, base_directory, curation, disease):
    """
    Télécharge le modèle et ses métadonnées, puis les enregistre dans un fichier zip.
    """
    try:
        model_id = model_data['id']
        sbml_url = model_data.get('url', None)
        title = model_data.get('name', "").lower()
        keywords = model_data.get('submitter_keywords', "").lower()

        if not sbml_url:
            print(f"Aucune URL trouvée pour le modèle {model_id}.")
            return

        # Créer le dossier s'il n'existe pas
        directory = os.path.join(base_directory, curation.replace(" ", "_"), disease)
        os.makedirs(directory, exist_ok=True)

        # Télécharger le fichier SBML
        model_path = download_model_file(model_id, sbml_url, directory, curation, disease)
        if model_path is None:
            return

        # Récupérer les métadonnées complètes
        try:
            full_metadata = s.get_model(model_id)
        except Exception as e:
            print(f"Erreur lors de la récupération des métadonnées complètes pour {model_id}: {e}")
            return

        # Sauvegarder les métadonnées en JSON
        metadata_filename = f"{model_id}_metadata.json"
        metadata_path = os.path.join(directory, metadata_filename)
        with open(metadata_path, 'w', encoding='utf-8') as f:
            json.dump(full_metadata, f, ensure_ascii=False, indent=4)

        # Créer un fichier zip contenant le modèle et ses métadonnées
        zip_filename = os.path.join(directory, f"{model_id}.zip")
        with zipfile.ZipFile(zip_filename, 'w') as zipf:
            zipf.write(model_path, os.path.basename(model_path))
            zipf.write(metadata_path, os.path.basename(metadata_path))

        # Supprimer les fichiers temporaires après zippage
        os.remove(model_path)
        os.remove(metadata_path)

        print("file", curation, disease, model_id) # jai changé ca aussi

    except Exception as e:
        print(f"Erreur lors du traitement du modèle {model_data['id']} : {e}")
```

```python
def main():
    base_directory = "downloaded_models"
    os.makedirs(base_directory, exist_ok=True)

    diseases = {
        "dengue": 'dengue',
        "west_nile": '"west nile"',
        "chikungunya": 'chikungunya',
        "mpox": '(mpox OR monkeypox)',
        "avian_influenza": '"avian influenza"',
        "covid": '(covid OR "SARS-CoV-2")'
    }
    curations = ['Manually curated', 'Non-curated']
    curations = ["Manually curated", "Non-curated"]

    for disease, disease_query in diseases.items():
        for curation in curations:
            query = f'{disease_query} AND curationstatus:"{curation}" AND modelformat:"SBML" AND TAXONOMY:9606'
            for m in get_all_models(query):
                download_model_with_metadata(m, base_directory, curation, disease)
if __name__ == "__main__":
    main()
```
